```python
import numpy as np

def clean_currency(value):
    if pd.isna(value):
        return np.nan
    # Use regex to replace '€', 'M', and 'K' while also handling '€ 1.2M' format
    value = str(value).replace('€', '').strip()
    if 'M' in value:
        return float(value.replace('M', '')) * 1_000_000
    elif 'K' in value:
        return float(value.replace('K', '')) * 1_000
    return float(value)

df['Value'] = df['Value'].apply(clean_currency)
df['Wage'] = df['Wage'].apply(clean_currency)
df['Release Clause'] = df['Release Clause'].apply(clean_currency)
```

```python
import pandas as pd
import numpy as np
import datetime

# Get the current year
current_year = datetime.datetime.now().year

# Create the new Contract Type column
conditions = [
    df['Contract'].str.contains('On Loan', case=False, na=False),
    df['Contract'].str.contains('Free', case=False, na=False),
    (df['Contract'].str.contains('~', na=False))
]
choices = ['On Loan', 'Free', 'Active']
df['Contract Type'] = np.select(conditions, choices, default=None)

# Initialize new year columns
df['Contract Start'] = None
df['Contract End'] = None

# Extract start and end years for 'Active' contracts
active_mask = df['Contract Type'] == 'Active'
contract_periods = df.loc[active_mask, 'Contract'].str.extract(r'(\d{4})\s*~\s*(\d{4})')
df.loc[active_mask, 'Contract Start'] = contract_periods[0]
df.loc[active_mask, 'Contract End'] = contract_periods[1]

# Extract end year for 'On Loan' contracts
on_loan_mask = df['Contract Type'] == 'On Loan'
loan_years = df.loc[on_loan_mask, 'Contract'].str.extract(r'(\d{4})\s*On Loan', expand=False)
df.loc[on_loan_mask, 'Contract End'] = loan_years

# Re-evaluate the 'Contract Type' based on the current year
# Now, filter out 'Active' contracts where the end year does not match the current year.
# The `fillna(-1)` is used to treat NaNs as a non-matching value.
df.loc[(df['Contract Type'] == 'Active') & (pd.to_numeric(df['Contract End'], errors='coerce') != current_year), 'Contract Type'] = 'Expired'

# Calculate the duration for 'Active' contracts
active_contracts_mask = df['Contract Type'] == 'Active'
df['Contract Duration'] = None  # Initialize the new column
duration = pd.to_numeric(df['Contract End']) - pd.to_numeric(df['Contract Start'])
df.loc[active_contracts_mask, 'Contract Duration'] = duration[active_contracts_mask]

# Convert the new year columns to numeric type
df['Contract Start'] = pd.to_numeric(df['Contract Start'])
df['Contract End'] = pd.to_numeric(df['Contract End'])

# Print the modified DataFrame to see the result
print(df[['Contract', 'Contract Type', 'Contract Start', 'Contract End', 'Contract Duration']])
```

```python
import pandas as pd
import numpy as np

# List the columns you want to transform
columns_to_convert = ['W/F', 'SM', 'IR']

# Loop through each column and perform the transformation
for column in columns_to_convert:
    # 1. Convert to string to ensure the .str accessor can be used
    df[column] = df[column].astype(str)
    
    # 2. Extract only the digits using a regular expression.
    # This safely handles cases with or without spaces before the star.
    df[column] = df[column].str.extract(r'(\d+)', expand=False)

    # 3. Convert to numeric using pd.to_numeric, with errors='coerce'
    # to handle any remaining non-convertible values by setting them to NaN.
    # Then convert to the nullable integer type 'Int64'.
    df[column] = pd.to_numeric(df[column], errors='coerce').astype('Int64')


```
