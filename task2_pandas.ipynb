```python
import pandas as pd
from textblob import TextBlob

def analyze_feedback(text):
    if pd.isna(text):
        return 'Neutral', 5.0
    
    # Create TextBlob object
    analysis = TextBlob(str(text))
    
    # Get polarity (-1 to 1)
    polarity = analysis.sentiment.polarity
    
    # Convert to 0-10 scale
    score = (polarity + 1) * 5
    
    # Categorize sentiment
    if score >= 7:
        sentiment = 'Good'
    elif score <= 3:
        sentiment = 'Bad'
    else:
        sentiment = 'Neutral'
    
    return sentiment, round(score, 1)

# Apply sentiment analysis
df[['sentiment', 'sentiment_score']] = df['feedback_text'].apply(
    lambda x: pd.Series(analyze_feedback(x))
)

# Display results
print(df[['feedback_text', 'sentiment', 'sentiment_score']].to_string(index=False))
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

def analyze_sentiment_vader(text):
    if pd.isna(text):
        return 'Neutral', 5.0
    
    scores = analyzer.polarity_scores(str(text))
    compound = scores['compound']  # -1 to 1
    
    # Convert to 0-10 scale
    score = (compound + 1) * 5
    
    # Categorize
    if compound >= 0.05:
        sentiment = 'Good'
    elif compound <= -0.05:
        sentiment = 'Bad'
    else:
        sentiment = 'Neutral'
    
    return sentiment, round(score, 1)

df[['sentiment', 'sentiment_score']] = df['feedback_text'].apply(
    lambda x: pd.Series(analyze_sentiment_vader(x))
)

print(df[['feedback_text', 'sentiment', 'sentiment_score']])
```

```python
import pandas as pd
from transformers import pipeline
import torch

# Initialize sentiment analysis pipeline
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-roberta-base-sentiment-latest",
    tokenizer="cardiffnlp/twitter-roberta-base-sentiment-latest"
)

def analyze_sentiment_transformers(text):
    if pd.isna(text):
        return 'Neutral', 5.0
    
    result = sentiment_pipeline(str(text))[0]
    label = result['label']
    confidence = result['score']
    
    # Map labels to scores
    if label == 'positive':
        sentiment = 'Good'
        score = 5 + (confidence * 5)  # 5-10 range
    elif label == 'negative':
        sentiment = 'Bad'
        score = 5 - (confidence * 5)  # 0-5 range
    else:  # neutral
        sentiment = 'Neutral'
        score = 5.0
    
    return sentiment, round(score, 1)

# Apply to first few rows (transformers can be slow)
df[['sentiment', 'sentiment_score']] = df['feedback_text'].head().apply(
    lambda x: pd.Series(analyze_sentiment_transformers(x))
)
```

```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download required NLTK data
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

def analyze_sentiment_nltk(text):
    if pd.isna(text):
        return 'Neutral', 5.0
    
    scores = sia.polarity_scores(str(text))
    compound = scores['compound']
    
    # Convert to 0-10 scale
    score = (compound + 1) * 5
    
    if compound >= 0.05:
        sentiment = 'Good'
    elif compound <= -0.05:
        sentiment = 'Bad'
    else:
        sentiment = 'Neutral'
    
    return sentiment, round(score, 1)

df[['sentiment', 'sentiment_score']] = df['feedback_text'].apply(
    lambda x: pd.Series(analyze_sentiment_nltk(x))
)
```
