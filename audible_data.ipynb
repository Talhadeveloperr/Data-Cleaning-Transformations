```python

df1['author'] = df1['author'].str.split(':').str[1]
print(df1['author'])
```

```python

df1['narrator'] = df1['narrator'].str.split(':').str[1]
print(df1['narrator'])
```

```python
df1['time'].head(10)
```

```python

hours = df1['time'].str.extract(r'(\d+)\s*hrs', expand=False).fillna(0).astype(int)
minutes = df1['time'].str.extract(r'(\d+)\s*mins', expand=False).fillna(0).astype(int)

# Calculate total minutes
df1['time'] = (hours * 60) + minutes

print(df1[['time']])
```

```python
df1.head(10)
```

```python
df1.drop(columns=['time_minutes'], inplace=True)
```

```python
df1['releasedate'] = pd.to_datetime(df1['releasedate'], dayfirst=True, format='%d-%m-%y')
print(df1['releasedate'])
```

```python
# Extract stars and ratings using regex
df1[['stars', 'ratings']] = df1['stars'].str.extract(r'([\d.]+) out of 5 stars(\d+) ratings?')

# Convert to numeric
df1['stars'] = pd.to_numeric(df1['stars'])
df1['ratings'] = pd.to_numeric(df1['ratings'])

df1[['stars', 'ratings']].head(10)
```

```python
# First, remove the commas
df1['price'] = df1['price'].str.replace(',', '')

# Then, use pd.to_numeric to convert, handling potential errors gracefully
df1['price'] = pd.to_numeric(df1['price'], errors='coerce')
```
