```python
import pandas as pd
```

```python
df=pd.read_csv('Uncleaned_DS_jobs.csv')
df.head(10)
```

```python
df.info()        # Summary of columns, data types, non-null counts
df.describe()    # Statistical summary (mean, std, etc.)
df.head()        # First few rows
df.tail()        # Last few rows
df.shape         # Number of rows and columns
df.columns 
```

Index

```python
df.drop(columns=['index'],inplace=True)
```

```python
df['Founded'].head(10)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')[0] + '-' + df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')[1]

df['Salary Estimate'].head(10)
```

###Company Name

```python
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
print(df['Company Name'])
```

##Founded

```python

df[['min_salary', 'max_salary']] = df['Salary Estimate'].str.split('-', expand=True).astype(int)
df['avg_salary'] = ((df['min_salary'] + df['max_salary']) / 2).astype(int)

# Display all columns
print(df[['Salary Estimate', 'min_salary', 'max_salary', 'avg_salary']].to_string(index=False))
```

```python
df.head(10)
```

```python
df.drop(columns=['Competitors'],inplace=True)
```

```python
import pandas as pd

# Extract state from Headquarters (last two characters after comma)
df['hq_state'] = df['Headquarters'].str.split(',').str[-1].str.strip()

# Extract state from Location (last two characters after comma)
df['job_state'] = df['Location'].str.split(',').str[-1].str.strip()

# Create same_state column (1 if states match, 0 otherwise)
df['same_state'] = (df['hq_state'] == df['job_state']).astype(int)

# Drop the temporary hq_state column if you don't need it
df = df.drop('hq_state', axis=1)

print(df[['Headquarters', 'Location', 'job_state', 'same_state']])
```

```python
reference_year = 2025
df['company_age'] = reference_year - df['Founded']
```
